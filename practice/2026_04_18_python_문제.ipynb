#Q1. 소수인지 판별해라.

* 숫자 하나 입력 -> 판별
* 소수인지 판별 = 약수가 1과 자기 자신--> 비교 대상: 1,자기 자신은 제외해도 됨.--> 2 ~ ( n-1) 비교해보자
* 약수인지 판별 = %가 0인 수 -> 약수 존재

> 변수 :
1. 입력되는 값의 변수
2. 결과를 알려주는 변수(소수가 맞다/ 아니다)


> 작업:
1. 데이터 가져오기(입력되는 변수)
2. 연산하기
3. 저장하기
4. 원하는 값을 화면에 출력


> 적용
1. 숫자가 2보다 큰지 판별
2. 1이 만족한다면 2~ (n-1) 정수 값을 i로 받음
3. i에서 숫자를 가져옴
4. n % i == 0 인지 판별
    4-1 맞다면 소수가 아님
    4-2 아니라면 다음 진행
5. 반복
