
###💻 Q2. 정수가 담긴 리스트 num_list가 주어질 때, num_list의 원소 중 짝수와 홀수의 개수를 담은 배열을 return 하도록 solution 함수를 완성해보세요.

```python
num_list = [1,2,3,4,5]
```

```python
num_list = [1,3,5,7]
```

```python
even = 0 # 초기화
odd = 0
for i in num_list:

    if i % 2 == 0: # 짝수
        even = even + 1
    elif i % 2 != 0:  # 홀수
        odd = odd + 1

print(even)
print(odd)
```

```python
print(even, odd)
```


###💻 Q5. 각에서 0도 초과 90도 미만은 예각, 90도는 직각, 90도 초과 180도 미만은 둔각 180도는 평각으로 분류합니다. 각 angle이 매개변수로 주어질 때 예각일 때 1, 직각일 때 2, 둔각일 때 3, 평각일 때 4를 return하도록 solution 함수를 완성해주세요.

```python
angle = 70
```

```python
angle = 91
```

```python
angle = 180
```

```python
# 예각 : 0 < angle < 90
# 직각 : angle = 90
# 둔각 : 90 < angle < 180
# 평각 : angle = 180
```

```python
if 0 < angle < 90:
    print("1")
elif angle == 90:
    print("2")
elif 90 < angle < 180:
    print("3")
elif angle == 180:
    print("4")
```
