문자열은 쪼갤 수 없어 join으로 뒤에 붙여야 어젯밤에 검색해봤다.

내 머릿속에서 바로 output되는건 append 인데  join 반갑다!??...?

for 로 하나씩 봐 그래서 aeiou 가 아니면 join해?



* 변수
    > 입력받는 문자열
    
    > 모음(제거할 문자들)

    > 출력할 문자열


* 작업
    > 입력받는 문자열을 하나씩 확인해. 쪼갤 수 없는데 어떻게 확인하지?(변환해)

    > '입력받는 문자열' 확인해서 aeiou가 아니면(if not) 출력할 문자열에 join 해

    > 이걸 '입력받은 문자열'이 끝날 때까지(=문자열의 길이만큼) 반복 작업(끝  몰라-> for n/ 반복 횟수 = len(문자열)


**    문제는 하나씩 문자열을 받아올 수 있냐다.**

```python
val = "aeiou"
output = []
```

```python
for 확인 in 문자열에서 하나씩 빼
    if(만약에) i 가 모음이 아니라면:
        입력받은 문자열.join

```

```python
for char in my_string:
    if char not in val:
        output.append(char)
result = "".join(output)

print(result)
```
