#🔍str.format()
lower() upper() join() split() find() replace() 예시 만들어보기


📝 str.format()
- 중괄호{} 를 구멍으로 사용하여, 문자열 뒤에 .format()을 붙여 구멍에 들어갈 값들을 넣어주는 방식
- 문자열과 입력 객체 정보의 조합으로 구성

💻   .format()

```python
x = 5; y = 10
print('The value of x is {} and y is {}'.format(x,y))
```

💻 .lower()_소문자로 변환

```python
X = "THIS WILL BE CHANGED TO SMALL LETTER."
lowercase_X = X.lower()
print(lowercase_X)
```

💻.upper()_대문자로 변환

```python
x = "this will be changed to capital letter."
uppercase_x = x.upper()
print(uppercase_x)
```

💻.join()_문자열 합치기
- 리스트 형식[], 튜플 형식(), 문자열 모두 가능하다.

```python
x =['apple','banana', 'carrot']
print("".join(x))        # 아무것도 없이 바로 연결
print(" ".join(x))       # " "(공란)한 칸 두고 연결
print("_".join(x))       # "_"(underspace) 두고 연결
```

```python
x =('apple','banana', 'carrot')
print("".join(x))        # 아무것도 없이 바로 연결
```

```python
x = "apple"
print("-".join(x))
```

💻.split()_문자열 합치기
- 리스트 형식[], 튜플 형식(), 문자열 모두 가능하다.

```python
x = "apple banana carrot"
print(x.split(" "))           # (" ")공란 기준으로 분해
print(x.split())              # () 기준으로 분해

x = "apple_banana_carrot"
print(x.split("_"))           # ("-") - 기준으로 분해
```

💻.find
- 문자열에서 특정 부분 문자열을 검색해서 처음 나타나는 인덱스를 반환
- 인덱스 0부터 시작

```python
x = "Let's find the word python."
print(x.find("python"))
print(x.find("find"))
```

💻.replace()
- 문자열 바꾸기

```python
x = "I am going to change word java to python"
print(x.replace("java","python"))
print(x.replace("python","java"))
```
