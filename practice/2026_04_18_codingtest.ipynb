(remove vs pop)
두 원소를 곱해야 하니
max를 한 번 더 이용해서 두 번째 큰값(현재로서는 가장 큰 값)을 max_2에 저장

````python
numbers.remove(max(numbers))
print(numbers)
````

````python
result = max_1 * max_2
print(result)
````
