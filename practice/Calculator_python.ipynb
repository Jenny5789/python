 소숫점 2자리까지 표기하자

숫자 리스트 item 개수 만큼 계산 반복 - > for i in range(len(numbers)): 사용


숫자 계산 반복 작업--> while = 까지 반복
예외조건 챙기기(0으로 나누기, 부호가 아닌거)


-  평균, 최댓값, 최솟값  구하기

```python
numbers = []

while True:
    user_input = input("숫자를 입력하세요(종료는 '='):     ")

    if user_input == "=":
        break

    try:
        num = float(user_input)
        numbers.append(num)

    except ValueError:
        print("숫자가 아닙니다. 숫자를 입력하세요")

print(f"입력받은 숫자: {numbers}")
```

```python
input_sign = input("부호를 입력하세요.( +, - , *, / ):   ")

result = numbers[0]
for i in range(1, len(numbers)):
    if input_sign == "+":
        result += numbers[i]

    elif input_sign == "-":
        result -= numbers[i]

    elif input_sign == "*":
        result *= numbers[i]

    elif input_sign == "/":
        result /= numbers[i]

    else:
        print("부호가 아닙니다. 부호를 입력하세요( +, - , *, / ):   ")

print(f"최종결과 : {result:.2f}")
```
