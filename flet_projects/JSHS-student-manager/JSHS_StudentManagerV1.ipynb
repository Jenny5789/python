#🏗️ JS HighSchool_StudentManager
    
> 🙋 JS 고등학생관리시스템 🙋
* 목적 :  '이름, 학년, 반, 번호'를 관리
* 기능 :
    - 이름별 - 학년/ 반/ 번호를 저장한다.
    - '이름' 검색 통해 학생정보를 조회가 가능하다.


        > 학생정보 수정/추가/삭제/출력이 가능하다.

## 🧐 변수

* 학생 이름: name (first_name + last_name)
* 학생 학년: grade
* 학생 반: class_num
* 학생 번호: num

* 수정할 값
* 출력은(이름 학년 반 번호) 조합으로 출력함


## 🧐 작업
* 저장되어 있는 last name + first name으로 이름을 만든다.
* 이름에 따라 각각 학년 반 번호를 만든 후 std_DB에 저장한다.
*





---


* 이름:  -> 'first name + last name' 조합으로 랜덤하게 학생수만큼 생성
    - 3개학년, 8개의 반, 번호는 25~30번까지 있다.(번호도 랜덤하게 설정한다)
    - for, random문을 이용하여 학생들의 이름을 생성한다.


* 학년: 랜덤함수(1~3학년)
* 반: 랜덤함수(1~8반)
* 번호: 랜덤함수(학생 수 만큼 부여한다/ 학생 고유 ID)(끝 번호가 25~30번)


````python
last_name = ["김", "이", "박", "최", "정", "강", "조", "윤", "장", "임"]
first_name = ["민준", "서준", "도윤", "주원", "예준", "현우", "준서", "지훈", "하준", "우진", "서연", "서윤", "지우", "서현", "민서",
    "하은", "지민", "윤서", "지안", "은서"]
````

````python
#이름 생성
import random as rand
# 이름 = last_name + first_name
name = rand.choice(last_name) + rand.choice(first_name)

print(name)
````

````python
# 1~3학년
for grade in range(1,4):
    print(grade)
````

````python
# 1~8반
for class_num in range(1, 9):
    print(class_num)
````

````python
# 마지막 번호 랜덤하게 받음( 25~ 30 으로 끝번호 받음)
last_std_num = rand.randint(25,30)
print(last_std_num)
````

- 학년 반 번호에 따른 이름 배정하기
    - 1학년 1반 마지막 번호(랜덤함수 이용)까지 이름 부여 작업 반복
    - 1학년 1~8반까지 8번 반복 작업
    - 2학년 3학년 동일


- 이름 부여 작업(= 반복작업이 가능한 요소)
```
    1 ~ 3 학년까지 반복,
    1 ~ 8 반까지 반복,
    번호만큼 반복(randint 받은 만큼)
    
```

````python
# 이름 받아 std_DB[] 에 저장
import random as rand
std_DB = []
for grade in range(1, 4):                  # 3학년까지 반복
    for class_num in range(1, 9):          # 8반까지 반복

        last_std_num = rand.randint(25, 30)  # 끝 번호 생성
        for num in range(1, last_std_num+1):   # 생성된 번호까지 반복

        # 이름 부여/ 이름 = last_name + first_name
            name = rand.choice(last_name) + rand.choice(first_name)
            std_info = {"이름":name, "학년": grade, "반":class_num, "번호": num}

            std_DB.append(std_info)

print(std_DB[:10])
print(f"전교 학생 수: {len(std_DB)}명")
````
