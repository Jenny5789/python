```python
# 숫자 추출
def extract_numbers(input_data):
    numbers = []         # 숫자로만 입력 받음
    not_numbers = []     # 리스트, 튜플--> 안에 확인 후 분류

    for item in input_data:   # 반복

        # isinstance(객체, 타입)  = 객체가 타입에 속하는지 확인하는 내장 함수

        if isinstance(item, (int, float)):      # 일반 숫자인지 확인
            numbers.append(float(item))

        elif type(item) in (list,tuple):          # list tuple 안에 있는 아이템 확인
            for sub_item in item:

                if isinstance(sub_item, (int, float)):
                    numbers.append(float(sub_item))  # 실수이면 뒤에 붙이기(numbers)
                else:
                    not_numbers.append(sub_item)     # 아니면 not_numbers로 분류
        # 숫자 아닌 것 분류
        else:
            not_numbers.append(item)

    # 최종 숫자 분류
    if not_numbers:
        print(f"입력된 데이터 중 숫자가 아닌 것들: {not_numbers}")
    return numbers

# 원시 데이터 입력받음
def get_raw_datas():
    raw_items = []
    print("=== 데이터를 입력하세요. (종료하려면 '=' 입력) ===")

    while True:
        user_input = input("입력:   ").strip()
        if user_input == "=":
            break

        # try except를 이용하여 1차 필터 내림
        try:
            num = float(user_input)
            raw_items.append(num)  # 숫자 변환해서 저장
        except ValueError:
            raw_items.append(user_input)  # 숫자 아니면 그대로 저

    return raw_items  # 필터링은 했지만 입력 받은 숫자 모두 받아냄


# 계산 실행
data = get_raw_datas()  # 사용자가 입력한 원시데이터 가져옴
final_numbers = extract_numbers(data)  # 숫자 추출함수에 넣어 진짜 숫자만 골라냄

# print(f"입력 받은 데이터: {data}")
# print(f"추출된 최종 숫자 리스트: {final_numbers}")

if not final_numbers:
    print("계산할 숫자가 없습니다. 종료합니다.")
else:
    print(f"추출된 최종 숫자:{final_numbers}")
    print("-" * 40)

    add_result = final_numbers[0]
    sub_result = final_numbers[0]
    mul_result = final_numbers[0]
    div_result = final_numbers[0]

    div_error = False

    for i in range(1, len(final_numbers)):
        num = final_numbers[i]

        add_result = add_result + num
        sub_result = sub_result - num
        mul_result = mul_result * num

        if num != 0:
            div_result /= num
        else:
            div_error = True

    print("-" * 40)
    print(f"더하기 결과: {add_result:.2f}")
    print(f"빼기 결과: {sub_result:.2f}")
    print(f"곱하기 결과: {mul_result:.2f}")

    if div_error:
        print(f"나누기 결과: 계산 중 0이 포함되어 나누기를 완료할 수 없습니다.")
    else:
        print(f"나누기 결과: {div_result:.2f}")


    print("-" * 40)
    avg = sum(final_numbers)/len(final_numbers)
    print(f"평균: {avg:.2f}")
    print(f"최댓값: {max(final_numbers):.2f}")
    print(f"최솟값: {min(final_numbers):.2f}")
    print("-" * 40)
```
