# 💻 Q1. 배열 원소의 길이
문자열 배열 strlist가 매개변수로 주어집니다. strlist 각 원소의 길이를 담은 배열을 return하도록 solution 함수를 완성해주세요.




```python
strlist = ["We", "are", "the", "world!"]
```

```python
strlist = ["I", "Love", "Programmers."]
```

* 문제
    - 리스트 안에 있는 각 원소의 길이를 구하라

* 변수
    - 입력받은 strlist(리스트)
    - 길이 구해 출력할 변수

* 작업
    - 리스트 안에 있는 리스트 인덱스별로 길이를 구한다.

```python
print(strlist[0])
```

```python
print(len("We"))
```

```python
print(len(strlist[0]))
```

주어진 리스트 안에 있는 strlist 의 인덱스에 따라 문자 길이를 구한다.

 동일하게 반복하면서 문자 길이 개수 결과를 새로운 리스트에 추가하여  붙여 나간다.

```python
result = []
for i in range(len(strlist)):
    result.append(len(strlist[i]))

print(result)
```

# 💻 Q2. 삼각형의 완성조건(1)
선분 세 개로 삼각형을 만들기 위해서는 다음과 같은 조건을 만족해야 합니다.

가장 긴 변의 길이는 다른 두 변의 길이의 합보다 작아야 합니다.
삼각형의 세 변의 길이가 담긴 배열 sides이 매개변수로 주어집니다. 세 변으로 삼각형을 만들 수 있다면 1, 만들 수 없다면 2를 return하도록 solution 함수를 완성해주세요.

```python
sides = [1, 2, 3]

```

```python
sides = [3, 6, 2]
```

```python
sides = [199,72,222]
```

'가장 긴 변의 길이 < 다른 두 변의 길이 합'

참일 경우 1 return,    거짓일 경우 2 return

* 변수
    - 입력받은 리스트
    - 가장 큰 변의 길이     ---->  x
    - 다른 두 변의 길이 합  ---->  x
    - 오름차순으로 정렬할 리스트 모음
    - 삼각형 완성조건 판단 여부

* 작업
    - 가장 큰 변의 길이가 무엇인지 파악한다.
        - a,b,c 오름차순 함수 이용하여 파악
    - 다른 두 변의 길이의 합을 구한다.
        - 나열한 리스트의 index 1+2--> 바로 구해
    - 삼각형의 조건을 만족하는지 판단한다.
        - if (index [0] < index[1] + index[2])
    - 판단한 결과를 출력한다.
        - T/F-->1/2

```python
# 오름차순 배열 함수 =
# sort() 원본의 순서를 직접 변경함. **반환값 None
# sorted() 원본은 그대로 두고, 새로운 리스트를 만들어 반환함
array = sorted(sides)
print(array)
```

```python
# 만약 (인덱스2의 값) <= (인덱스 0 + 인덱스1) 이면
#   참이면 1 출력 거짓이면 2 출력


if array[2] < (array[0] + array[1]):
    print("1")
else:
    print("2")
```
