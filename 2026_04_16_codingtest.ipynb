###💻 Q2.  정수 n이 주어질 때, n이하의 짝수를 모두 더한 값을 return 하도록 solution 함수를 작성해주세요.

```python
n = 10
```

```python
n = 4
```

10 이하 짝수 --> n % 2 == 0 (나머지  = 0 = 짝수)--> 더해

if 짝수이면 더해

홀수이면 -1 해서 짝수 선별 후 더해

근데 숫자는 어떻게 알지?....
1부터 n 까지 하나씩 뽑아 n%2 대입해-> 나머지 =0 그럼 더해 / 아니면 +1
끝을 아는 반복 - > for

```python
sum = 0 # 초기화
for i in range(1, n+1):  # n까지 확인 필요--> +1
    if i % 2 == 0:
        sum= sum+i
    else:
        i = i+1
print(sum)
```

```python
n = 100
```

두 숫자의 곱이 n --> 약수 구하기?

(1,n ) ......(n,1)    두 쌍은 무조건 포함

+1로 >n stop     /   (x +1) % == 0 그럼 있어

```python
20 % 1 == 0
```

```python
count = 0
for i in range(1,n+1): # range(n+1) 부터 하면 0부터 시작되니 , 시작점인 1 도 지정
    if n % i == 0:  # 나머지 0
        count= count + 1
    else:
        i = i + 1
print(count)
```

몇 사분면인지 구하기-->

```python
print(dot[0])
print(dot[1])
```

```python
if dot[0] > 0:  # x축 양수
    if dot[1] > 0:
        print("1")  # y 축 양수
    else:
        print("4")  # y 축 음수

elif dot[0] < 0:    # x 축 음수
    if dot[1] > 0:
        print("2")  # y 축 양수
    else:
        print("3")  # y 축 음수
```
