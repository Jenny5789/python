### 🧐
> 테스트 (프레임 ->  중간점검 -> 첫번째 투구, 두번째 투구)
    
    프레임 예외 처리
    
    첫 번째 투구, 두 번재 투구(= 10 - 첫 번째  넘어진 핀 개수)

    점수 계산 함수에 첫 번째 투구, 두 번재 투구


> 점수 계산 함수

    첫 번째 시도 -> 10개 ----------------> 10 +  다음 두 번의 투구의 결과

    첫 번째 두 번째 시도 -> 10개 --------> 10 +  다음 한 번의 투구의 결과

    첫 번째 두 번째 시도 -> 10개 안됨  --> 첫 번째, 두 번째 투구의 결과의 합

> 변수

    프레임 당 쓰러진 핀의 수  [,]
    
    다음 투구의 결과를 더하는 횟수 : 0 ~ 2
    
    프레임 당 점수 계산 완료 여부 = T/F
