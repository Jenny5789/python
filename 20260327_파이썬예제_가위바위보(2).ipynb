#👩‍🏫문제] 컴퓨터와 가위바위보 게임을 코딩해 보세요. 5전 3승제로 먼저 이기는 사람을 알려주고, 축하의 인사도 출력해 보세요



> '20260325_파이썬예제_가위바위보(1).ipynb'  중복 제거 및 함수로 묶어 정리



```python
#사용자 컴퓨터 게임횟수 0으로 초기화 설정
winner_user, winner_computer, total_game = 0, 0, 0
total_game = winner_user + winner_computer

# List로 RSP 값을 숫자로 설정, DIct로 key값 부여하여 경기 결과 사전 설정
list_RSP = {0:'R', 1:'S', 2:'P' }
dict_RSP_judge = {
    'R':{'R':None,'S':'user','P':'computer'},
    'S':{'R':'computer','S':None,'P':'user'},
    'P':{'R':'user','S':'computer','P':None}
}

# 사용자 RSP 입력
while True:           # 이건 왜 들어가야지??????????????
    user_RSP = None
    while True:
        user_RSP = input("R(주먹) S(가위) P(보) 중 하나 선택해서 입력해주세요.: ")
        if user_RSP in ['R', 'S', 'P']:
            break
        print("다시 입력해주세요.")

# 컴퓨터 랜덤 불러오고 RSP값 받기
    import random
    com_RSP = list_RSP[random.randint(0,2)]
    print("컴퓨터가 선택한 것은 {}입니다.".format(com_RSP))

 # 승부 결정
    winner = dict_RSP_judge[user_RSP][com_RSP]

# 게임횟수 누적
    if winner == 'user':
        winner_user = winner_user + 1
        total_game = total_game + 1
        print("{}가 승리했어요.(경기횟수: {})".format('user',total_game))
    elif winner == 'computer':
        winner_computer = winner_computer + 1
        total_game = total_game + 1
        print("{}가 승리했어요.(경기횟수: {})".format('computer',total_game))
    else:
        print("무승부예요. 다시 한번 해보세요.")

    if total_game >=5:
        break
```
