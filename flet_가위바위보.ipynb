👩‍🏫문제] 컴퓨터와 가위바위보 게임을 코딩해 보세요. 5전 3승제로 먼저 이기는 사람을 알려주고, 축하의 인사도 출력해 보세요.
-( 무승부는 다시 경기 진행!)

```python
import random
def pick_RSP(who):
    RSP = None
    if who == "user":
        while True:
            user_RSP = input("R(주먹) S(가위) P(보) 중 하나 선택해서 입력해주세요: ").upper()
            if user_RSP in ['R', 'S', 'P']:
                return user_RSP
            print("다시 입력해주세요.")
    else:
        com_RSP = list_RSP[random.randint(0,2)]
        return com_RSP

def winner_count(user_RSP, com_RSP, score):
    winner = dict_RSP_judge[user_RSP][com_RSP]

    if winner == 'user':
        score['user'] = score['user'] + 1
        score['total'] = score['total'] + 1
        print("사용자가 승리했어요. 축하합니다! (현재 {}전 {}승)".format(score['total'],score['user']))

    elif winner == 'computer':
        score['com'] = score['com'] + 1
        score['total'] = score['total'] + 1
        print("컴퓨터가 승리했어요! (현재 {}전 {}승)".format(score['total'],score['com']))

    else:
        score['total'] = score['total'] + 1
        print("무승부예요. (현재 {}전)".format(score['total']))


    return score

score = {'user':0,'com':0, 'total':0}

list_RSP = {0:'R', 1:'S', 2:'P'}
dict_RSP_judge = {
    'R':{'R':None,'S':'user','P':'computer'},
    'S':{'R':'computer','S':None,'P':'user'},
    'P':{'R':'user','S':'computer','P':None}
    }

while True:
    user_RSP = pick_RSP('user')
    com_RSP = pick_RSP('computer')
    score = winner_count(user_RSP, com_RSP, score)

    if score['total'] >= 5:
        print("GAME OVER (5판 완료)\n")

        if score['user'] > score['com']:
            print(f"최종결과: 사용자가 {score['user']}승으로 최종 승리했습니다!\n")
        elif score['com'] > score['user']:
            print(f"최종결과: 컴퓨터가 {score['com']}승으로 최종 승리했습니다!\n")
        else :
            print(f"최종결과: {score['user']}: {score['com']}으로 무승부입니다!\n")

        break
```
