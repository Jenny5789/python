```python
class PhoneBook:
    def __init__(self):
        self.phone_book = {}

    def add_contact(self):
        name = input("이름을 입력하세요: ")
        number = input("전화번호를 입력하세요: ") or None
        email = input("메일주소를 입력하세요: ") or None
        job = input("직장명을 입력하세요: ") or None

        self.phone_book[name] = {
            "전화번호": number,
            "이메일": email,
            "직업": job
        }
        print(f"{name}님의 정보가 저장되었습니다.\n")

    def search_contact(self):
        name = input("검색할 이름을 입력하세요: ")
        if name in self.phone_book:
            info = self.phone_book[name]
            print(f"\n--- {name}님 정보 ---")
            print(f"전화번호: {info['전화번호']}")
            print(f"이 메 일: {info['이메일']}")
            print(f"직    업: {info['직업']}")
        else:
            print("해당 이름의 정보가 없습니다.")

    def delete_contact(self):
        name = input("삭제할 이름을 입력하세요: ")
        if name in self.phone_book:
            del self.phone_book[name]
            print(f"{name}님의 정보를 삭제했습니다.")
        else:
            print("등록되지 않은 이름입니다.")

    def list_contacts(self):
        print("\n--- 전체 연락처 목록 ---")
        if not self.phone_book:
            print("저장된 연락처가 없습니다.")
        else:
            for name, info in self.phone_book.items():
                print(f"이름: {name} | 전화번호: {info['전화번호']} | 이메일: {info['이메일']} | 직업: {info['직업']}")

    def run(self):
        """프로그램 실행 루프"""
        while True:
            print("\n--- 전화번호부 메뉴 ---")
            print("1. 정보 추가")
            print("2. 정보 검색")
            print("3. 정보 삭제")
            print("4. 전체 목록 보기")
            print("5. 종료")

            menu = input("원하는 작업 번호를 입력하세요: ")

            if menu == "1":
                self.add_contact()
            elif menu == "2":
                self.search_contact()
            elif menu == "3":
                self.delete_contact()
            elif menu == "4":
                self.list_contacts()
            elif menu == "5":
                print("프로그램을 종료합니다.")
                break
            else:
                print("잘못된 선택입니다. 다시 입력하세요.")

if __name__ == "__main__":
    app = PhoneBook()
    app.run()
```
