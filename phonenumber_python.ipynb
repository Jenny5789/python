Q. 전화번호부 _ 개인 이름, 이메일 주소, 전화번호, 직장명

사용자가 개인정보 알려주면 검색 통해 특정 정보 제공

1. 이름, 전화번호, 이메일 주소, 직장명 을 입력받는다.

     이름 전화번호 이메일주소 직장명을 각각 input 이용하여 입력받아
     입력 받은 데이터들을 리스트로 저장?
     없으면 None

2. 이름을 키로 검색하여 기타 정보를 알 수 있게 딕셔너리 형태로 저장한다.

    리스트에 저장된 데이터들을 키값은 '이름',
    
    나머지는 '이름 딕셔너리' 안에 또 다른 딕셔너리로 저장

3. 여러 개의 정보를 어디에 저장할래?


4. 검색을 어떻게 해서 어떻게 받을래?


```python
name = input("이름을 입력하세요: ")
number = input("전화번호를 입력하세요:")
email = input("메일주소를 입력하세요: ")
job = input("직장명을 입력하세요:")
정보 = f"이름: {name}, 전화번호:{number}, 메일주소:{email}, 직업:{job}"

print(정보)
#print(f"이름: {name}, 전화번호:{number}, 메일주소:{email}, 직업:{job}")
```

이름을 입력받고 이어서 (전화번호 메일주소 직업을) 정보를 반복해서 입력 받음-->  while 문 사용 받음

```python
while True:
    name = input("이름을 입력하세요: ")
    if name == "종료":
        break
    else:
        number = input("전화번호를 입력하세요:")
        email = input("메일주소를 입력하세요: ")
        job = input("직장명을 입력하세요:")
        print(f"{name}님의 정보는, {number}, {email}, {job}로 저장되었습니다.")
```
