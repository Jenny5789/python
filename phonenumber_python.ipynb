1. 이름, 전화번호, 이메일 주소, 직장명 을 입력받는다.

     이름 전화번호 이메일주소 직장명을 각각 input 이용하여 입력받아

이름을 입력받고 이어서 (전화번호 메일주소 직업을) 정보를 반복해서 입력 받음-->  

while 문 사용 받음



빈 전화번호부에 저장, 데이터 입력값 없으면 None에 저장, 키 값은 이름 --> 이름으로 저장되었습니다.

```python
phone_book = {}
while True:
    name = input("이름일 입력하세요: (종료 시 '종료' 입력) ")
    if name == "종료":
        print("입력을 종료합니다. 저장된 전체 목록: ", phone_book)
        break

    else:
        number = input("전화번호를 입력하세요:") or None
        email = input("메일주소를 입력하세요: ") or None
        job = input("직장명을 입력하세요:") or None

        phone_book[name] = {
            "전화번호" : number, "이메일" : email, "직업" : job
        }
        print(f"{name}님의 정보가 저장되었습니다. \n")
```

 검색을 어떻게 해서 어떻게 받을래?---> 이것도 딕셔너리 키값을 이용해서 검색

이름을 입력해서 검색해

저장한 phonebook에서 이름을 검색해 정보를 뽑아와/ 없으면 정보가 없다고 알려줘

```python
name = input("이름을 입력하세요: ")
if name in phone_book:
    info = phone_book[name]
    print(f"--- {name}님 정보 ---")
    print(f"전화번호: {info['전화번호']}")
    print(f"이 메 일: {info['이메일']}")
    print(f"직    업: {info['직업']}")
else:
    print("해당 이름의 정보가 없습니다.")
```

삭제---> 이름을 입력하게 한다.
이름을 phone_book에서 찾아 정보를 삭제한다.
이름이 없으면 등록되지 않은 이름이라고 한다.

```python
name = input("삭제할 이름을 입력하세요: ")
if name in phone_book:
    del phone_book[name]
    print(f"{name}님의 정보를 삭제했습니다.")
else:
    print("등록되지 않은 이름입니다.")
```

종합하자.
1. 데이터 입력
2. 검색
3. 삭제
4. 종료

if - elif -else 구조. 1234 각각 번호에 따라 동작하도록 함

```python
phone_book = {}

while True:
    print("\n--- 전화번호부 메뉴 ---")
    print("1. 정보 추가")
    print("2. 정보 검색")
    print("3. 정보 삭제")
    print("4. 전체 목록 보기")
    print("5. 종료")

    menu = input("원하는 작업 번호를 입력하세요: ")

    # 1. 정보 추가
    if menu == "1":
        name = input("이름을 입력하세요: ")
        number = input("전화번호를 입력하세요: ") or None
        email = input("메일주소를 입력하세요: ") or None
        job = input("직장명을 입력하세요: ") or None

        phone_book[name] = {
            "전화번호": number,
            "이메일": email,
            "직업": job
        }
        print(f"{name}님의 정보가 저장되었습니다. \n")

    # 2. 정보 검색
    elif menu == "2":
        name = input("검색할 이름을 입력하세요: ")
        if name in phone_book:
            info = phone_book[name]
            print(f"--- {name}님 정보 ---")
            print(f"전화번호: {info['전화번호']}")
            print(f"이 메 일: {info['이메일']}")
            print(f"직    업: {info['직업']}")
        else:
            print("해당 이름의 정보가 없습니다.")

    # 3. 정보 삭제
    elif menu == "3":
        name = input("삭제할 이름을 입력하세요: ")
        if name in phone_book:
            del phone_book[name]
            print(f"{name}님의 정보를 삭제했습니다.")
        else:
            print("등록되지 않은 이름입니다.")

    # 4. 전체 목록 보기
    elif menu == "4":
        print("\n--- 전체 연락처 목록 ---")
        if not phone_book:  # 딕셔너리가 비어있는지 확인
            print("저장된 연락처가 없습니다.")
        else:
            for name, info in phone_book.items():
                print(f"이름: {name} | 전화번호: {info['전화번호']} | 이메일: {info['이메일']} | 직업: {info['직업']}")

    # 5. 종료
    elif menu == "5":
        print("프로그램을 종료합니다.")
        break

    else:
        print("잘못된 선택입니다. 다시 입력하세요.")
```

함수로 묶어보기

```python
def add_contact(phone_book):
    """연락처 추가 기능"""
    name = input("이름을 입력하세요: ")
    number = input("전화번호를 입력하세요: ") or None
    email = input("메일주소를 입력하세요: ") or None
    job = input("직장명을 입력하세요: ") or None

    phone_book[name] = {
        "전화번호": number,
        "이메일": email,
        "직업": job
    }
    print(f"{name}님의 정보가 저장되었습니다.\n")


def search_contact(phone_book):
    """연락처 검색 기능"""
    name = input("검색할 이름을 입력하세요: ")
    if name in phone_book:
        info = phone_book[name]
        print(f"\n--- {name}님 정보 ---")
        print(f"전화번호: {info['전화번호']}")
        print(f"이 메 일: {info['이메일']}")
        print(f"직    업: {info['직업']}")
    else:
        print("해당 이름의 정보가 없습니다.")


def delete_contact(phone_book):
    """연락처 삭제 기능"""
    name = input("삭제할 이름을 입력하세요: ")
    if name in phone_book:
        del phone_book[name]
        print(f"{name}님의 정보를 삭제했습니다.")
    else:
        print("등록되지 않은 이름입니다.")


def list_contacts(phone_book):
    """전체 목록 출력 기능"""
    print("\n--- 전체 연락처 목록 ---")
    if not phone_book:
        print("저장된 연락처가 없습니다.")
    else:
        for name, info in phone_book.items():
            print(f"이름: {name} | 전화번호: {info['전화번호']} | 이메일: {info['이메일']} | 직업: {info['직업']}")


def main():
    """메인 실행 함수"""
    phone_book = {}

    while True:
        print("\n--- 전화번호부 메뉴 ---")
        print("1. 정보 추가")
        print("2. 정보 검색")
        print("3. 정보 삭제")
        print("4. 전체 목록 보기")
        print("5. 종료")

        menu = input("원하는 작업 번호를 입력하세요: ")

        if menu == "1":
            add_contact(phone_book)
        elif menu == "2":
            search_contact(phone_book)
        elif menu == "3":
            delete_contact(phone_book)
        elif menu == "4":
            list_contacts(phone_book)
        elif menu == "5":
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 선택입니다. 다시 입력하세요.")

if __name__ == "__main__":
    main()
```
