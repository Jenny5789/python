##❓ 문제:사용자로부터 숫자를 입력 받아 내림차순, 오름차순으로 출력하세요.

```python
a = [5,9,4,1,3,8,7,]
```

- 💻 내림차순

```python
for i in range(0,len(a)-1,1):
    for j in range(i+1, len(a)):
        if a[i] < a[j]:
            n = a[i]
            a[i] = a[j]
            a[j] = n
print(a)
```

- 💻 오름차순

```python
for i in range(0,len(a)-1,1):
    for j in range(i+1, len(a)):
        if a[i] > a[j]:
            n = a[i]
            a[i] = a[j]
            a[j] = n
print(a)
```

###🔹 종류_int, float, complex, type, isinstance, Decimal, Fraction
- int(): 정수
- float(): 소수
- complex(): 복소수
- type() : 클래스의 데이터 형을 확인
- isinstance() : 주어진 값이 특정 클래스에 속하는지를 확인

###🔹 종류_Mathematics: math
- ceil(x): (천장함수) 입력된 숫자보다 크거나 같은 가장 작은 정수를 반환하는 올림 함수, 무조건 올림하여 정수로 올림
- floor(x): (바닥함수) 입력된 실수보다 크지 않은 최대 정수를 반환하는 내림함수. 소수점 아래를 무조건 버
- exp(x): 지수함수, e^x
- pow(x,y): x^y
- sqrt(x): 루트 x
- pi: 파이함수(3.141592...)
- e: 자연상수 e함수 (2.71828....)

```python
import math
print(math.ceil(1.15))
print(math.ceil(1.95))
# 천장함수, 입력된 값 양수 -> 반올림
print(math.ceil(-1.85))
print(math.ceil(-1.35))
# 천장함수, 입력된 값 음수 -> 반림
print(math.ceil(0))
print(math.ceil(3))
print(math.ceil(-3))
# 천장함수, 입력된 값 정수 -> 정수 그대로 출력
```

```python
import math
print(math.floor(1.15))
print(math.floor(1.95))
# 천장함수, 입력된 값 양수 -> 반내림
print(math.floor(-1.85))
print(math.floor(-1.35))
# 천장함수, 입력된 값 음수 -> 반내림
print(math.floor(0))
print(math.floor(3))
print(math.floor(-3))
# 천장함수, 입력된 값 정수 -> 정수 그대로 출력
```

```python
import math
print(math.exp(3))      # e^3
print(math.pow(5,2))    # 5^2
print(math.sqrt(5))     # 루트 5
print(math.pi)          # pi의 값
print(math.e)           # e의 값
```

###🔹 종류_Mathematics: random

###🔹 List

- 여러 요소를 한번에 작업할 수 있는 데이터 유형
- 요소를 변경할 수 있음
- 요소를 추가할 수 있음
    - append(): 리스트 맨 뒤에 하나의 아이템이 추가됨.
    - extend(): 리스트 맨 뒤에 여러 개의 아이템이 추가됨
    - insert(위치, 추가할 data): 특정 위치에 아이템 추가
    - 결합, 연결 지원'+' 이용
    - 반복 수행 지원 '*'

- 요소 제거 가능
    - del(): 아이템, 변수도 제거, 연결고리를 끊어줌
    - remove(): 첫번째로 만나는 값을 삭제
    - pop(): 뽑아온다
    - clear(): 빈 리스트로 초기화. 변수는 남아있고 아이템이 삭제되는 거다.

- slicing(특정부분), indexing(아이템)을 이용하여 요소를 추가 제거함.

```python
my_list=['g','o','h','o','m','e',['h','e','l','l','o']]
```

```python
print(my_list[:])     # 전체 출력
print(my_list[1:3])   # index 1번째부터 3이전(=2번째)까지 출력
print(my_list[-2:])   # 리스트 뒤부터 index '-1'로 설정. -2번째부터 끝까지 출력
print(my_list[2:12])  # index 2부터 11까지(=12-1)이지만, index는 6까지만 있다.--> 6까지만 출력
print(my_list[-1:])   # index -1부터 끝까지 출력 = 마지막만 출력
print(my_list[6][-1]) # index -1부터 끝까지 출력
print(my_list[-1:])   # index -1부터 끝까지 출력 = 마지막만 출력
print(my_list[6][-1]) # index 6 안에 index -1 출력
print(my_list[-2:-4]) # index -2부터 -4까지로 적혀있지만 출력 방향이 다름.
# 아무것도 못 담고 빈리스트[]를 반환


# 슬라이싱의 규칙: 왼쪽에서 오른쪽으로 가면서 아이템을 담는다.
```

- 원 데이터 수정
    - append(): 뒤에 하나 추가
    - extend(): 뒤에 여러 개 추가
    - insert(): 특정 위치에 추가
    - remove():삭제
    - clear(): 빈 리스트로 초기화
    - sort(): 정렬
    - reverse(): 앞뒤 뒤집기

- 결과 전달 받음
    - pop(): 꺼내받기
    - index(): 위치 알려줌
    - count(): 개수 세어줌
    - copy(): 똑같이 가짜 리스트 줌

       - shallow copy(): 주소를 동일하게 저장. 리스트 수정 시 데이터 같이 변함,(바로가기)
       - deep copy(): 모든 item에 대해 주소를 새로 저장해준다.(복사본)

```python
my_list = [3,2,1,7,9,6,4]
a = 1
my_list.append(a)



print(my_list.pop(4))
print(my_list.index(1))
print(my_list.count(5))
```
