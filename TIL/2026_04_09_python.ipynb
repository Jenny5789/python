#📚 Object & Class

##📝 Object-oriented programming(OOP)_객체 지향 프로그래밍

#🔹 객체 지향 프로그래밍

*   다중 패러다임 프로그래밍 언어 _ 고유한 특성과 각각의 특성을 하나로 묶어 객체 중심으로 설계
*   재사용 가능한 코드 작성에 초점 _ 3대 요소: 상속, 캡슐화, 다형성
*   class는 속성과 동작에 대해 정의함 _ 변수(자동차 색상)와 함수(자동차 주행하기)로 이루어져있다
*   Object는 class의 인스턴스(객체)라 함.

        **class**: 정의. 자동차
        **object**: 실제 물건, 빨간차
        (예: 클래스= 사과, 객체: 사과1, 사과2, 사과3)


````python
#💻
class Human:
    species = "human"

    def __init__(self,name, age):
        self.name = name
        self.age = age

kim = Human("kim", 10)
lee = Human("lee", 15)

print("kim is a {}".format(kim.__class__.species))
print("lee is also a {}".format(lee.__class__.species))

print("{} is {} years old".format(kim.name, kim.age))
print("{} is {} years old".format(lee.name,lee.age))

print("{} is {} years old".format(lee.__class__.species,lee.age))
````

#🔹 Method

*   함수에 정의되어 있는 함수(클래스 내부 소속)
*   객체(인스턴스)가 주체가 되어 호출(예: 객체.메소드)
*   객체의 속성을 활용한 동작
*   **'(self)'**을 받아 주체가 나(객체 자신)임을 가리킴


````python
#💻
# class attribute(동일한 특성, 이 사람은 인간이다.)
class Human:

    # instance attribute(각각의 특성, 이 사람은 한국인이다/ 50살이다.)
    def __init__(self,name,age):
        # __init__: 생성자라고 하며, 초기화 시키고 자동호출된됨(필수로 적어야한다!)
        # (self): 내가 만들어 내가 사용한다. 전지적 시점
        self.name = name
        self.age = age

    # instance method(각각의 동작, 이 사람은 말을 한다/ 밥을 먹는다)
    def sing(self,song):
        return"{} sings {}".format(self.name,song)
    def dance(self):
        return"{} is now dancing".format(self.name)

# instaniate the Parrot class
kim = Human("kim",10)

# call our instance methods
print(kim.sing("'Happy'"))
print(kim.dance())
````

#🔹 상속(Inheritance)___ Class의🌸

* 기본 클래스의 내용을 수정하지 않고, 그대로 사용하기 위해 자식 클래스를 생성하는 방식

* 많은 사람들이 이용할 수 있고, 필요에 따라 수정 변경 가능
* '부모클래스/자식클래스'로 이루어짐

* 부모클래스: 물려주는 클래스/ 자식클래스: 물려받는 클래스

* 자식클래스에 선언되어 있지 않아도 부모클래스 함수를 사용 가능함

* 공통적인 것은 부모클래스로 정의하고, 상속 받은것(각각의 특성)은 자식 클래스로 정의함



```
# 부모 클래스
class Bird:

# 자식 클래스- 'class(부모클래스)' 형태로 미리 선언 해아함
class Penguin(Bird):

#  자식클래스에서 사용하면 부모클래스를 호출 가능함.
super().__init__()    - 부모가 미리 설정해둔 속성들을 그대로 가지고와 자식들이 덧붙일 수 있다.

```


```
A = 부모함수()    # A = 부모클래스 생성
A = 자식함수()    # A = 자식클래스 생성

A.부모함수()      # A.야 나는 자식이지만, 부모클래스 함수 사용할게
A.자식함수()      # A.야 자식클래스 함수 사용할게
```



````python
# parent class
class Bird:
    def __init__(self):           # 자기 자신 외 필요한 매개변수가 없다(self)
        print("Bird is ready")
    def whoisThis(self):
       print("Bird")
    def swim(self):
        print("Swim faster")

# child class
class Penguin(Bird):
    def __init__(self):

        # call super() function
        super().__init__()
        print("Penguin is ready")
    def whoisThis(self):
        print("Penguin")
    def run(self):
        print("Run faster")

peggy = Penguin()     # 자기 자신 외 필요한 매개변수가 없다(self)
peggy.whoisThis()     # peggy 본인클래스인, 자식클래스에서 이용함
peggy.run()           # peggy 본인클래스인,자식클래스에서 이용함
peggy.swim()
# peggy 본인클래스이 없으니,'class Penguin(Bird):' 'super().__init__()'을 이용해 부모클래스 이용함
````

#🔹 캡슐화(Encapsulation)

*   Method 변수에 대한 접근 제한
*   직접적인 데이터 수정 방지
*   class에서 만드는 것 자체는 **외부**에서 변경 접근이 X.
*   class에서 만드는 것 자체는 **내부**에서 변경 접근이 O.

>  '_ _변수 _ _' (생성자 함수): 초기 세팅, 객체 만들면 자동 호출됨
>
> '_ _변수' (비공개 변수) : 접근 제한 및 보호, 필요 시 직접 호출







````python
class Computer:
    def __init__(self):
        # class 생성하면서 '__init__' 자동 호출
        self.__maxprice = 900
        # '__maxprice'비공개 변수, 접근 제한 및 보호

    def sell(self):
        print("SellingPrice:{}".format(self.__maxprice))
        # class 내부에서 '.format'을 이용하여 비공개 변수를 {} 안에 넣어 읽어줄게

    def setMaxPrice(self,price):
        self.__maxprice = price
        # class 내부에서 비공개 변수('__maxprice') 정의했으니 접근 및 수정할 수 있게 해줄게

c = Computer()
c.sell()
# 자기 자신 외 필요한 매개변수가 없다(self)

#change the price
c.__maxprice = 1000
c.sell()
# 비공개 변수가 있어 내부의 진짜 값은 변경할 수 없다.

#using setter function
c.setMaxPrice(1000)
c.sell()
# 'setMaxPrice'(본인)로 들어가 함수를 변경할 수 있다.
````

#🔹 다형성(Polymorphism)

*   다양한 데이터 유형에 대한 공통 인터페이스(=공통된 조작 방법,함수)를 사용
* 동일한 명령을 내려도, 객체에 따라 적용되는 방식만 달리 작용됨

````python
class Parrot:
    def fly(self):  # Parrot 의 fly 유형
        print("Parrot can fly")
    def swim(self):
        print("Parrot can't swim")

class Penguin:
    def fly(self):   # Penguin 의 fly 유형
        print("Penguin can't fly")
    def swim(self):
        print("Penguin can swim")

# common interface(공통 동작)
def flying_test(bird):
    bird.fly()
    # 'fly()' 공통

#instantiate objects
blu = Parrot()
peggy = Penguin()

# passing the object
flying_test(blu)    # Parrot() 유형 실행
flying_test(peggy)  # Penguin() 유형 실행
````

#🔹 상속_isinstance(), issubclass()

````python
class Polygon:
    def __init__(self,no_of_sides):
        self.n=no_of_sides
        self.sides=[0 for i in range(no_of_sides)]
    def inputSides(self):
        self.sides=[float(input("Enterside"+str(i+1)+":")) for i in range(self.n)]
    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])
````

````python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self,3)
        #super().__init__(3)

def findArea(self):
    a,b,c=self.sides
    #calculate the semi-perimeter
    s=(a+b+c)/2
    area=(s*(s-a)*(s-b)*(s-c))**0.5
    print('The area of the triangle is %0.2f'%area)

t=Triangle()
t.inputSides()
t.dispSides()
t.findArea()
````

    - ex. __init__, __str__ 등


#🔹 특수함수(special functions)

*  '_ _변수' (비공개 변수) : 접근 제한 및 보호, 필요 시 직접 호출
*  연산자(+)는 특수함수(__add__)를 호출하기 위한 편의도구다.
* 의도하지 않아도 필요할 때 알아서 동작하는 함수다.




````python
class Point:
    def __init__(self,x=0,y=0):
        # __init__ 생성될 때 자동 생성
        self.x=x
        self.y=y

    def __str__(self):
         # __str__ 출력될 때 자동 생성
        return"({0},{1})".format(self.x,self.y)

    def __add__(self,other):
        # __add__ 더해질 때 자동 생성
        x = self.x+other.x
        y = self.y+other.y
        return Point(x,y)

p1=Point(1,2)
p2=Point(2,3)

print(p1+p2)     #연산자(+) 역할
print(Point.__add__(p1,p2))
````
