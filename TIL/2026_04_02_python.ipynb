- 물품명에 대해 단가와 재고를 랜덤하게 설정 후 DB로 설정해보자
-  모든 데이터에 대해' 단가 & 재고'값을 설정해야 함-> 반복 작업-->  for 사용

```python
import random as rand

물품정보들 = {}   # 초기화
for i in range(len(물품명)):
    단가 = rand.randrange(1000, 20000,1000)
    재고 = rand.randint(50,100)
    물품정보들[물품명[i]] = {"단가": 단가,"재고":재고}
```

```python
def 물품정보생성(물품명):
    import random as rand

    물품정보들 = {}   # 초기화
    for i in range(len(물품명)):
        단가 = rand.randrange(1000, 20000,1000)
        재고 = rand.randint(50,100)
        물품정보들 = [물품명[i]] = {"단가": 단가,"재고":재고}
    return 물품정보들

print(물품정보들)
```

- print(물품정보들)한 결과 가로로 나열된다.--> for in 사용하여 세로로 정리해보자

```python
for item in 물품정보들:
    print("item:", item, 물품정보들[item])
```
