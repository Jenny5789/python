[데이터 생성 예시]
초등학교 1~ 6학년이 있어요. 각 학년마다 반이 5~8반이 있어요.
각 반에는 20~30명의 학생이 있어요
시험 과목은 " 영어", "국어", "수학", "과학"이 있어요.

각 학생들의 모든 과목에 대한 성적을 하나의 변수에 저장하세요.

```python
초등학교 = ["1학년", "2학년", "3학년", "4학년", "5학년", "6학년"]
```

```python
초등학교[0]
print(초등학교[0])
```

```python
학년 = {}
for i in 초등학교:
    학년[i] = {}
print(학년)
```

```python
학생 = {초등학교[0]: {"반":None, "번호": None}}
print(학생)
```

```python
import random as rand
반 = rand.randint(1,8)         # 1반부터 8반까지
번호 = rand.randint(20,30)     # 20명~ 30명
print(반)
print(번호)
```

```python
import random as rand
학생 = {}
for i in range(len(초등학교)):
    반 = rand.randint(1,8)
    번호 = rand.randint(20,30)
    학생[초등학교[i]] = {"반":반,"번호":번호}
```

```python
print(학생)
```
