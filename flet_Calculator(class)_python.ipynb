```python
import re

class Calculator:
    # 핵심: data_list=None 이라고 적어서, 데이터를 안 넣어도 오류가 나지 않게 함
    def __init__(self, data_list=None):
        if data_list is None:
            self.raw_data = []
        else:
            self.raw_data = data_list

        self.numbers = []
        self.non_numbers = []

    # 데이터 입력받기 (이미 데이터가 있어도 추가 가능)
    def get_input(self):
        print("=== 데이터를 입력하세요. (종료하려면 '=' 입력) ===")
        while True:
            user_input = input("입력: ").strip()
            if user_input == "=":
                break

            try:
                num = float(user_input)
                self.raw_data.append(num)
            except ValueError:
                self.raw_data.append(user_input)

    # 숫자 추출
    def extract_numbers(self):
        # 기존 데이터를 초기화하지 않고, 새로 추출할 때마다 새로 비움
        self.numbers = []
        self.non_numbers = []

        for item in self.raw_data:
            # 1. 이미 숫자라면 그대로 통과
            if isinstance(item, (int, float)):
                self.numbers.append(float(item))

            # 2. 리스트나 튜플이라면 (기존 로직 유지)
            elif isinstance(item, (list, tuple)):
                for sub_item in item:
                    if isinstance(sub_item, (int, float)):
                        self.numbers.append(float(sub_item))
                    else:
                        self.non_numbers.append(sub_item)

            # 3. 문자열이라면! (여기서 숫자를 찾아내기)
            elif isinstance(item, str):
                # r"[-+]?\d*\.\d+|\d+"는 '숫자를 찾아라'라는 패턴입니다.
                match = re.search(r"[-+]?\d*\.\d+|\d+", item)
                if match:
                    self.numbers.append(float(match.group()))
                else:
                    self.non_numbers.append(item)

            # 4. 그 외에는 숫자가 아님
            else:
                self.non_numbers.append(item)

        if self.non_numbers:
            print(f"입력된 데이터 중 숫자가 아닌 것들: {self.non_numbers}")

    # 숫자 출력 포맷을 정리하는 도우미 메서드
    def _format_value(self, value):
        # 1. 소수점 둘째 자리까지 반올림 (부동소수점 오차 방지)
        rounded_val = round(value, 2)

        # 2. 정수인지 확인
        if rounded_val == int(rounded_val):
            return str(int(rounded_val)) # 정수면 정수 부분만 출력
        else:
            return f"{rounded_val:.2f}"  # 아니면 소수점 2자리까지 출력

    def display_statistics(self):
        """계산 결과를 수행하고 출력합니다."""
        if not self.numbers:
            print("계산할 숫자가 없습니다.")
            return

        print(f"추출된 최종 숫자: {self.numbers}")
        print("-" * 40)

        add_result = sub_result = mul_result = div_result = self.numbers[0]
        div_error = False

        for num in self.numbers[1:]:
            add_result += num
            sub_result -= num
            mul_result *= num

            if num != 0:
                div_result /= num
            else:
                div_error = True

        print(f"더하기: {self._format_value(add_result)} | 빼기: {self._format_value(sub_result)}")
        print(f"곱하기: {self._format_value(mul_result)}")

        if div_error:
            print("나누기: 0이 포함되어 계산 불가")
        else:
            print(f"나누기: {self._format_value(div_result)}")

        print("-" * 40)
        print(f"평균: {self._format_value(sum(self.numbers)/len(self.numbers))}")
        print(f"최댓값: {self._format_value(max(self.numbers))} | 최솟값: {self._format_value(min(self.numbers))}")
        print("-" * 40)

# --- 실행부 ---
if __name__ == "__main__":

    calc = Calculator()
    calc.get_input()


    calc.extract_numbers()
    calc.display_statistics()
```
